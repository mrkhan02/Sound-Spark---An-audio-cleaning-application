```python
#importing the necessary modules
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import wavfile
#reading the .wav file, nrate is the sampling rate, data contains all the samples
#nrate for the given wav file is 16,0000 samples per second
#the duration of the audio is 4 seconds and it has a total of 64,000 samples
nrate, data = wavfile.read(r'C:\\Users\Dell\Downloads\a2\test_noise.wav')
#print(ndata)
#plotting the samples
plt.plot(ndata)
plt.show()
```

```python
#converting all the samples to list form
data=data.tolist()
#applying a padding of 4 zeroes to the left so that we do not get nan values for the first 4 samples
pad=4
for i in range(0,pad):
    data.insert(0,0)

#creating a list which contains 64,004 zeroes
newdata=[0]*64004

#we are considering a rectangular window where the ith sample will be the average of the previous 5 samples including the ith sample
for i in range(4,64004):
    newdata[i]=(data[i]+data[i-1]+data[i-2]+data[i-3]+data[i-4])/5

#removing the padded values so that we get our final samples
mov=newdata[4:]
#converting the list into the 16 bit encoding(for wav file)
mov = np.array(mov, dtype=np.int16)
#writing to the wav file
wavfile.write('test1a.wav',nrate,mov)
#plotting the cleaned signal, we conclude that the signal has been made more smooth
plt.plot(mov)
plt.show()
```
